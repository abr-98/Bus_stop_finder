# Modelling

```python
class Neural_Network:
    def __init__(self,training,name,X_train,y_train,X_test,y_test,s,training_columns):
        self.name=name
        self.scalar=s
        if training:
            self.model,self.history=Get_trained_model_on_data(training_columns,
                                                              X_train,y_train,
                                                              X_test,y_test,
                                                              data_class=self.name)
        else:
            self.model=tf.keras.models.load_model('./saved/{}_model.h5'.format(self.name))

        self.barplots,self.lineplots=plot_all_information(self.model,
                                                          self.name,
                                                          X_test,y_test)
        
        self.train_acc,self.test_acc,self.classify_rept=get_numeric_reports(self.model,
                                                                            X_train,y_train,
                                                                            X_test,y_test)
        clear_output()
        
    def predict_classes(self,X):
        return np.round(self.model.predict(X))
        
    def summary(self):
        self.model.summary()
```

```python
class Merged_Neural_Network:
    def __init__(self,name='Project_model',training=False):
        self.bus_stop_model=Neural_Network(training,'bus_stop',
                                           X_train_bus_stop,y_train_bus_stop,
                                           X_test_bus_stop,y_test_bus_stop,
                                           s_bus_stop,
                                           [16, 13, 9, 1, 10, 5, 0, 6, 14, 15, 7])
        
        self.norm_road_model=Neural_Network(training,'norm_road',
                                            X_train_norm_road,y_train_norm_road,
                                            X_test_norm_road,y_test_norm_road,
                                            s_norm_road,
                                            [1, 0, 15, 9, 11, 14, 10, 2, 7, 5, 6, 13, 16, 12])
        self.Signal_model=Neural_Network(training,'Signal',
                                         X_train_Signal,y_train_Signal
                                         ,X_test_Signal,y_test_Signal,
                                         s_Signal,
                                         [3, 4, 1, 0, 10, 15, 2, 9, 8, 14, 11, 13, 7, 5, 12, 6, 16])
        self.Turn_model=Neural_Network(training,'Turn',
                                       X_train_Turn,y_train_Turn,
                                       X_test_Turn,y_test_Turn,
                                       s_Turn,
                                       [4, 1, 10, 2, 0, 9, 8, 6, 11, 16, 13, 12, 15, 5, 7])
        self.Congestion_model=Neural_Network(training,'Congestion',
                                             X_train_Congestion,y_train_Congestion,
                                             X_test_Congestion,y_test_Congestion,
                                             s_Congestion,
                                             [0, 2, 11, 8, 10, 9, 1, 12, 13, 16, 6, 15, 5, 7])
                                             
        if training:
            print('Trained')
        else:
            print('Loaded')
        
    def predict(self,X):
        bsp=self.bus_stop_model.predict_classes(self.bus_stop_model.scalar.transform(X))
        nrp=self.norm_road_model.predict_classes(self.norm_road_model.scalar.transform(X))
        sp=self.Signal_model.predict_classes(self.Signal_model.scalar.transform(X))
        tp=self.Turn_model.predict_classes(self.Turn_model.scalar.transform(X))
        cp=self.Congestion_model.predict_classes(self.Congestion_model.scalar.transform(X))
        return np.concatenate([bsp,nrp,sp,tp,cp],axis=1)
```

```python
project_model=Merged_Neural_Network(training=False)
```
